### 4. Локализация занятости по разным секторам экономики Российской Федерации и г. Москвы:

```python
HEADER = 'Локализация занятости по разным секторам экономики Российской Федерации и г. Москвы с 2015 по 2020 гг.'
```

```python
data = pd.read_csv('ЗСЭ+.csv')
data.drop(columns=data.columns[0], inplace=True)
```

```python
moscow_data = data[data['Регион'] == 'г. Москва'].reset_index(drop=True)
russia_data = data[data['Регион'] == 'Российская Федерация'].reset_index(drop=True)

division_df = moscow_data.copy()
for column in moscow_data.columns[2:]:
    division_df[column] = (moscow_data[column] / russia_data[column]) * 100

data_viewer.display_data(header=HEADER, data=division_df)
```

```python
data_viewer.plot_ratios(data=division_df, header=HEADER)
```
