### 1. Динамика производительности труда Российской Федерации и г. Москвы:

```python
HEADER = 'Динамика производительности труда Российской Федерации и г. Москвы за 2010–2020 гг.'
```

```python
vrp_data = pd.read_excel('ВРП.xlsx', engine='openpyxl')
cz_data = pd.read_excel('ЧЗ.xlsx', engine='openpyxl')
```

```python
common_columns = vrp_data.columns.intersection(cz_data.columns)
numeric_columns = [col for col in common_columns if pd.api.types.is_numeric_dtype(vrp_data[col]) and pd.api.types.is_numeric_dtype(cz_data[col])]

result_data_numeric = vrp_data[numeric_columns].div(cz_data[numeric_columns])
result_data = pd.concat([vrp_data[common_columns[0]], result_data_numeric], axis=1).reset_index(drop=True)

data_viewer.display_data(header=HEADER, data=result_data)
```

```python
data_viewer.plot_data(data=result_data, mode='all', header=HEADER)
```

### 5. Динамика обеспеченности бюджета г. Москвы налоговыми и неналоговыми доходами:

```python
HEADER = 'Динамика обеспеченности бюджета г. Москвы налоговыми и неналоговыми доходами за 2014–2021 гг.'
data_viewer.FOOTER = 'Рассчитано по данным Казначейства России (Росказна).'
```

```python
data = pd.read_excel('ННД.xlsx', engine='openpyxl')
```

```python
data_base_year_normalized = data.copy()

for idx, row in data.iterrows():
    if row['Регион'] != 'Российская Федерация':
        data_base_year_normalized.iloc[idx, 1:] = (row.iloc[1:] / row.iloc[1]) * 100

data_without_country = data_base_year_normalized[data_base_year_normalized['Регион'] != 'Российская Федерация']

data_viewer.display_data(header=HEADER, data=data_without_country)
```

```python
data_viewer.plot_data(data=data_base_year_normalized, mode='region', header=HEADER)
```

### 6. Динамика обеспеченности бюджета г. Москвы безвозмездными поступлениями:

```python
HEADER = 'Динамика обеспеченности бюджета г. Москвы безвозмездными поступлениями за 2014–2021 гг.'
```

```python
data = pd.read_excel('БП.xlsx', engine='openpyxl')
```

```python
data_base_year_normalized = data.copy()

for idx, row in data.iterrows():
    if row['Регион'] != 'Российская Федерация':
        data_base_year_normalized.iloc[:, 1:] = data_base_year_normalized.iloc[:, 1:].astype(int)

data_without_country = data_base_year_normalized[data_base_year_normalized['Регион'] != 'Российская Федерация']

data_viewer.display_data(header=HEADER, data=data_without_country)
```

```python
data_viewer.plot_data(data=data_base_year_normalized, mode='region', header=HEADER)
```
